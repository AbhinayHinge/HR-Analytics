# **ML MODEL ON employee Attrition dataset**

As we see some good correlation between LastYearsPromoted and Attrition so we perform ML model on df2 dataset

# Logistic Regression

SYNTAX:

glm( formula , data , family)

```r
head(df2)
```

```r
# Select required columns for regression
input <- df2[ , c("MonthsInOrg","LastPayIncrementBand","LastPromotionYears","Attrition")]
print(head(input))
```

```r
table(df2$Attrition)
```

230 Employees left the job

770 Employees not left the job

```r
dim(input)
```

```r
# Split into Training and Testing

train_input = input[1:800, ]
test_input = input[801:1000, ]
```

```r
# To get dimension of train data
dim(train_input)
```

```r
# View the input value of train data
View(train_input)
```

```r
# Dimension of test data
dim(test_input)
```

```r
Attrition.data <- glm( formula = Attrition~ MonthsInOrg+LastPayIncrementBand+LastPromotionYears, data = input, family = binomial)

print(Attrition.data)
```

p value of Last promotion years is greater than 5 that means variable which is important for prediction is last promotion years

```r
# Predict values on test data
Attrition.pred= predict(Attrition.data, newdata = test_input, type = "response")
print(Attrition.pred)
```

```r
# Predict yes no
Attrition.pred_yes_no = ifelse(Attrition.pred > 0.5, 1, 0)
Attrition.pred_predicted <- factor(Attrition.pred_yes_no, levels = c(0,1))
print(Attrition.pred_predicted)
```

```r
# Predicted values
m.pred_actual <- test_input$Attrition
m.pred_actual
```

```r
# compare predicted and actual values
Attrition.pred_predicted == m.pred_actual
```

All values are true that means our prediction is correct

```r
sum(Attrition.pred_predicted == m.pred_actual)
```

We tested for 200 values and we got all of them TRUE i.e. model is accurate

```r
mean(Attrition.pred_predicted == m.pred_actual)
```

we got accuracy of 100% that means all predictions made are true

```r
table(Attrition.pred_predicted, m.pred_actual)
```

Here confusion matrix explains actual and predicted values here there are 43 employees who left job and our model predicted it correctly
